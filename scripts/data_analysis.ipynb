```python
short_kernel_it = []
avg_range_short = []
avg_range_mid = []
mid_anchors_total = []
mid_larger_than_64 = []
mid_larger_than_1024 = []
long_anchors_total = []
long_larger_than_64 = []
long_larger_than_1024 = []
short_anchors_total = []
short_larger_than_64 = []
short_larger_than_1024 = []

def read_csv_files_range_dis(range_file, mid_range_file, long_range_file):
    with open (range_file, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        header = next(reader)
        range_row = next(reader)
            
    # Convert numerical values to integers (excluding headers)
    ranges_numeric = [int(x) for x in range_row[2:]]
    # print(ranges_numeric[0:10])

    with open(mid_range_file, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        header = next(reader)
        mid_range_row = next(reader)
        
    # Convert numerical values to integers (excluding headers)
    mid_ranges_numeric = [int(x) for x in mid_range_row[2:]]
    # print(mid_ranges_numeric[0:100])

    with open(long_range_file, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        header = next(reader)
        long_range_row = next(reader)
        
    # Convert numerical values to integers (excluding headers)
    long_ranges_numeric = [int(x) for x in long_range_row[2:]]
    # print(long_ranges_numeric[0:100])

    short_ranges_numeric = [ x - y - z for x, y, z in zip(ranges_numeric, mid_ranges_numeric, long_ranges_numeric)]
    
    sum_short_kernel_it = 0
    for i in range(0, len(short_ranges_numeric)):
        sum_short_kernel_it += short_ranges_numeric[i]* (-(-i // 64))
    
    short_kernel_it.append(sum_short_kernel_it/sum(short_ranges_numeric))
    # print("Precent of Anchors > 16 bases: ", sum(ranges_numeric[16:]) / sum(ranges_numeric))
    # print("Precent of Anchors > 16 bases in short kernels", sum(short_ranges_numeric[16:]) / sum(short_ranges_numeric))
    # print("Precent of Anchors > 64 bases in short kernels", sum(short_ranges_numeric[64:]) / sum(short_ranges_numeric))
    # print("Precent of Anchors > 16 bases in mid kernels", sum(mid_ranges_numeric[16:]) / sum(mid_ranges_numeric))
    # print("Precent of Anchors > 64 bases in mid kernels", sum(mid_ranges_numeric[64:]) / sum(mid_ranges_numeric))
    # print("Precent of Anchors in Long ", sum(long_ranges_numeric) / sum(ranges_numeric))
    # print("Precent of Anchors in Mid ", sum(mid_ranges_numeric) / sum(ranges_numeric))
    
    
    
    workload_range_dis  = [value * index for index, value in enumerate(ranges_numeric)]
    workload_long_range_dis = [value * index for index, value in enumerate(long_ranges_numeric)]
    workload_mid_range_dis = [value * index for index, value in enumerate(mid_ranges_numeric)]
    workload_short_range_dis = [x - y - z for x, y, z in zip(workload_range_dis, workload_mid_range_dis, workload_long_range_dis)]
    
    avg_range_short.append(sum(workload_short_range_dis) / sum(short_ranges_numeric))
    avg_range_mid.append(sum(workload_mid_range_dis) / sum(mid_ranges_numeric))
    
    mid_anchors_total.append(sum(mid_ranges_numeric))
    mid_larger_than_64.append(sum(mid_ranges_numeric[64:]))
    mid_larger_than_1024.append(sum(mid_ranges_numeric[1024:]))
    long_anchors_total.append(sum(long_ranges_numeric))
    long_larger_than_64.append(sum(long_ranges_numeric[64:]))
    long_larger_than_1024.append(sum(long_ranges_numeric[1024:]))
    short_anchors_total.append(sum(short_ranges_numeric))
    short_larger_than_64.append(sum(short_ranges_numeric[64:]))
    short_larger_than_1024.append(sum(short_ranges_numeric[1024:]))
```
